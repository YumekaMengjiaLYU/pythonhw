```python
from sklearn.preprocessing import StandardScaler  
from sklearn.preprocessing import MinMaxScaler  
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Takes the initial iterate with the first component = mu, the second component = sigma
# The second argument is the dataset.
from sklearn.datasets import load_boston
boston_dataset = load_boston()
X = boston_dataset.data
y = boston_dataset.target
boston = np.array(boston_dataset.data)
print(boston)
std_scaler = StandardScaler()
mm_scaler = MinMaxScaler()
X = std_scaler.fit_transform(X)
#boston = (boston - np.mean(boston, axis = 0)) / np.std(boston, axis = 0)
print(boston)
y = mm_scaler.fit_transform(y.reshape(-1, 1))
print(X)
print(y)
print(X.shape[1])
print(y.shape[0])
```

```python

def gradient_descent_ridge_reg(X, y, l2_penalty, num_epochs=100, learning_rate = 0.01):
    w = np.zeros((X.shape[1],1))
    b = 0
    N = y.shape[0]
    global l_history 
    l_history  = []
    global w_norm_history 
    w_norm_history = []
    global b_history
    b_history = []
  
    for i in range(num_epochs-1):
        
        temp = y-(b+np.dot(X, w))
        w_partial = -2*np.matmul(X.T,temp)/N + 2*l2_penalty*w
        b_partial = 2*np.sum(temp)/N
        w_norm= np.linalg.norm(w)
        w_norm_history += [w_norm]
        l = 1/N*np.sum((y-(b+np.dot(X, w)))**2) + l2_penalty*((np.linalg.norm(w))**2)
        l_history += [l] 
        b_history+= [b]
        b_new = b + learning_rate * b_partial 
        w_new = w - learning_rate * w_partial
        b = b_new
        w = w_new
        
        
        
    return w_norm


```

```python
gradient_descent_ridge_reg(X, y, l2_penalty=0.1, num_epochs=100, learning_rate = 0.01)
```

```python
print(len(l_history), len(w_norm_history))

num_epochs = np.linspace(0, 100, num = 100)

# plotting the line 1 points  
plt.scatter(num_epochs, l_history, marker = "*", label = "reg objective value") 
  
# line 2 points 

# plotting the line 2 points  
plt.scatter(num_epochs, w_norm_history, marker = "+", label = "2-norm of the weight") 
  
# naming the x axis 
plt.xlabel('Epoch Number') 
# naming the y axis 
plt.ylabel('Value') 
# giving a title to my graph 

  
# show a legend on the plot 
plt.legend() 

plt.savefig('4c.png', bbox_inches='tight')
# function to show the plot 
plt.show() 
```
